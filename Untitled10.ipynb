```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = yf.download('TSLA' , start='2018-01-01' , end= "2025-01-01")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df.tail()
```

```python
df['Next_Close'] = df['Close'].shift(-1)
```

```python
df['MA_5'] = df['Close'].rolling(5).mean()
df = df.dropna()
```

```python
X = df[['Open', 'High', 'Low', 'Volume', 'MA_5']]
y = df['Next_Close']
```

```python
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
lr = LinearRegression().fit(X_train, y_train)
rf = RandomForestRegressor(n_estimators=50, random_state=42).fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)
```

```python
print(f"Linear Regression R²: {r2_score(y_test, lr_pred):.4f}")
print(f"Random Forest R²:     {r2_score(y_test, rf_pred):.4f}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(y_test.values, label='Actual', color='black')
plt.plot(lr_pred, label='Linear Reg', color='blue', alpha=0.7)
plt.plot(rf_pred, label='Random Forest', color='red', alpha=0.7)
plt.title('TSLA: Actual vs Predicted')
plt.legend()
plt.savefig('prediction.png', dpi=150)
plt.show()
```
